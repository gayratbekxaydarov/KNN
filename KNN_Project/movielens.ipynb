```python
import pandas as pd

movies = pd.read_csv('/home/gayratbek/university/leetcode/KNN_Project/movies.csv')
ratings = pd.read_csv('/home/gayratbek/university/leetcode/KNN_Project/ratings.csv')

# Filmlar va reytinglarni movieId orqali birlashtiramiz
movie_data = pd.merge(ratings, movies, on='movieId')
print(movie_data)
```

```python
# Pivot table yaratamiz
movie_features_df = movie_data.pivot_table(index='title', columns='userId', values='rating').fillna(0)
print(movie_features_df.head())
```

```python
from scipy.sparse import csr_matrix

# Matritsani kichraytiramiz
movie_features_matrix = csr_matrix(movie_features_df.values)
print(movie_features_matrix)
```

```python
movie_counts = movie_data.groupby('title')['rating'].count()

popular_movies = movie_counts[movie_counts >= 10].index
filtered_data = movie_data[movie_data['title'].isin(popular_movies)]

movie_features_df = filtered_data.pivot_table(index='title', columns='userId', values='rating').fillna(0)
print(movie_features_df)
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

movie_features_matrix = csr_matrix(movie_features_df.values)

model_knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20)
model_knn.fit(movie_features_matrix)
```

```python
def get_recommendations(movie_name):
    if movie_name not in movie_features_df.index:
        print(f"'{movie_name}' topilmadi. Iltimos, nomini tekshiring.")
        return
    movie_idx = movie_features_df.index.get_loc(movie_name)
    
    # KNN orqali eng yaqin 6 ta qo'shnini topamiz
    distances, indices = model_knn.kneighbors(
        movie_features_df.iloc[movie_idx, :].values.reshape(1, -1), 
        n_neighbors=6
    )
    
    print(f"'{movie_name}' uchun tavsiyalar:\n")
    for i in range(1, len(distances.flatten())):
        recommended_movie = movie_features_df.index[indices.flatten()[i]]
        print(f"{i}: {recommended_movie} (Distance: {distances.flatten()[i]:.3f})")
get_recommendations('Toy Story (1995)')
```
